```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa_data = pd.read_csv("../input/datasetfifa/fifa.csv", index_col = "Date")
fifa_data.head()
```

### this data can be visualized meaningfully through categorical plots only we'll use the stripplot layered ontop of a violin plot to do so

```python
plt.figure(figsize = (16,6))
sns.stripplot(data = fifa_data, marker = 'X')
```

```python
plt.figure(figsize = (16,6))
sns.violinplot(data = fifa_data, color = 'gray')
sns.stripplot(data = fifa_data, marker = 'X')
```

```python
spotify_data = pd.read_csv("../input/spotifyx/spotifyx.csv", index_col = "Date")
spotify_data.head()
```

### **this dataset also is categorical but we can use lineplot bc there's not to many categories, but count plot helps us understand the quantity better**

```python
plt.figure(figsize = (16,6))
sns.lineplot(data = spotify_data.tail())
```

```python
plt.figure(figsize = (16,6))
sns.countplot(data = spotify_data, facecolor = (0,0,0,0), linewidth = 5, edgecolor=sns.color_palette("spring", 5))
```

```python
flight_delays_data = pd.read_csv("../input/flightdelays/flight_delays.csv", index_col = "Month")
flight_delays_data.head()
```

### **this dataset has alot of categories and each one has 2 pairs so to use a barplot I've taken only one category**

```python
plt.figure(figsize = (16,6))
plt.title("Avg arrival delays by month")
sns.barplot(x = flight_delays_data.index, y = 'NK', data = flight_delays_data, palette = 'spring')
```

### **to visualize the whole dataset in way other is by using a heatmap**

```python
plt.figure(figsize = (16,6))
sns.heatmap(flight_delays_data, linewidths = 1, cmap = 'summer', annot = True)
```

```python
insurance_data = pd.read_csv("../input/data-visualizatiion/insurance.csv")
insurance_data.head()
```

### **this dataset has a lot of mixed values and types of values so we can use most type of plots on it so as you see i'll be using alot of plots on this data set**

```python
plt.figure(figsize = (25,10))
sns.scatterplot(data = insurance_data, x = 'bmi', y = 'charges', hue = 'sex')
```

```python
sns.relplot(data = insurance_data, 
            x = 'age', y = 'charges',
            style = 'smoker',
            kind = 'scatter',
           )
```

```python
sns.relplot(data = insurance_data, 
            x = 'age', y = 'charges',
            style = 'smoker',
            kind = 'scatter',
            col = 'sex'
           )
```

```python
plt.figure(figsize = (25,10))
sns.jointplot(data = insurance_data, 
            x = 'age', y = 'charges',
            kind = 'hist',
           )
```

```python
plt.figure(figsize = (25,10))
sns.violinplot(data = insurance_data, x = 'children', y = 'charges', 
               hue = 'sex', split = True, palette = 'spring')
```

```python
plt.figure(figsize = (50,40))
g = sns.pairplot(insurance_data, kind = 'hist')
```
